```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('QueryResults.csv',names=['DATE','TAG','POSTS'],header=0)
df.head()
```

```python
#Each row is a date, each column is a prog langauge. Store under a variable called reshaped_df
reshaped_df = df.pivot(index='DATE',columns='TAG',values='POSTS')
```

```python
reshaped_df.shape
```

```python
reshaped_df.head()
```

```python
reshaped_df.count()
```

```python
#substitute the NULL values with 0 for each NaN value
reshaped_df.fillna(0,inplace=True)
reshaped_df = reshaped_df.fillna(0)
reshaped_df.head()
```

```python
reshaped_df.isna().values.any()
```

```python
#plot the popularity of the Java programming language
#Also change the size - width(16),height(10)
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date',fontsize=14)
plt.ylabel('Number Of Posts',fontsize=14)
plt.ylim(0,35000)
plt.plot(reshaped_df.index,reshaped_df.java,reshaped_df.python)
```

```python
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date',fontsize=14)
plt.ylabel('Number Of Posts',fontsize=14)
plt.ylim(0,35000)
#Plot all languages using for loop
for column in reshaped_df.columns: 
  plt.plot(reshaped_df.index,reshaped_df[column],
           linewidth=3,label=reshaped_df[column].name)
  
plt.legend(fontsize=16)
```

```python
# The window is number of observations that are averaged
roll_df = reshaped_df.rolling(window=6).mean()
plt.figure(figsize=(16,10))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Date',fontsize=14)
plt.ylabel('Number Of Posts',fontsize=14)
plt.ylim(0,35000)

#Plot the roll_df instead
for column in roll_df.columns: 
  plt.plot(roll_df.index,roll_df[column],
           linewidth=3,label=roll_df[column].name)
  
plt.legend(fontsize=16)
```
